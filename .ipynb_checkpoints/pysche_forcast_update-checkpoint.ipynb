```python
#Load .env variable(s)

load_dotenv()
```

```python
#set up Alpaca API key and secret key

alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")
```

```python
#Create the alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key, 
    alpaca_secret_key,
    api_version="v2"
)
```

```python
#Format current data as ISO format
start_date = pd.Timestamp("2021-08-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2022-08-01", tz="America/New_York").isoformat()
```

```python
#Set the tickers
psyche_tickers = ["MNMD","CYBN","ATAI","CMPS","PSYK"]
           
timeframe = "60Min"   
```

```python
#Get 1 years data for psychedelic stocks/etf

psychedelics_stocks = alpaca.get_bars(
    psyche_tickers,
    timeframe,
    start = start_date,
    end = end_date
).df

#Display dataframe

display(psychedelics_stocks.head())
```

```python
#Reorganize Dataframe
#Separate ticker data

MNMD = psychedelics_stocks[psychedelics_stocks["symbol"]== "MNMD"].drop("symbol",axis = 1)
CYBN = psychedelics_stocks[psychedelics_stocks["symbol"]== "CYBN"].drop("symbol",axis = 1)
ATAI = psychedelics_stocks[psychedelics_stocks["symbol"]== "ATAI"].drop("symbol",axis = 1)
CMPS = psychedelics_stocks[psychedelics_stocks["symbol"]== "CMPS"].drop("symbol",axis = 1)
PSYK = psychedelics_stocks[psychedelics_stocks["symbol"]== "PSYK"].drop("symbol",axis = 1)



#Concatenate DataFrames

mind_trip_portfolio = pd.concat([MNMD,CYBN,ATAI,CMPS], axis =1 , keys = ["MNMD","CYBN","ATAI","CMPS"])

#View DataFrame 

display(mind_trip_portfolio.head())
```
